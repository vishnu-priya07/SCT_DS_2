```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="whitegrid")
titanic_data = pd.read_csv('train.csv')
print("First 5 Rows of the Dataset:")
print(titanic_data.head())
print("\n Dataset Info:")
print(titanic_data.info())
print("\n Summary Statistics:")
print(titanic_data.describe(include='all'))
print("\n Missing Values Before Cleaning:")
print(titanic_data.isnull().sum())
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
titanic_data.drop(columns=['Cabin'], inplace=True)
categorical_cols = ['Survived', 'Pclass', 'Sex', 'Embarked']
for col in categorical_cols:
    titanic_data[col] = titanic_data[col].astype('category')
print("\n Missing Values After Cleaning:")
print(titanic_data.isnull().sum())
sns.countplot(x='Survived', data=titanic_data)
plt.title('Survival Count')
plt.show()
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.show()
sns.countplot(x='Survived', hue='Pclass', data=titanic_data)
plt.title('Survival Rate by Passenger Class')
plt.show()
sns.histplot(titanic_data['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
sns.countplot(x='Embarked', hue='Survived', data=titanic_data)
plt.title('Survival Rate by Port of Embarkation')
plt.show()
sns.histplot(titanic_data['Fare'], bins=40, kde=True)
plt.title('Fare Distribution')
plt.show()
plt.figure(figsize=(8, 6))
sns.heatmap(titanic_data.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
sns.boxplot(x='Survived', y='Age', data=titanic_data)
plt.title('Age vs. Survival')
plt.show()
print("\n Survival Rates by Group:")
print(titanic_data.groupby('Sex')['Survived'].apply(lambda x: x.astype(int).mean()))
print(titanic_data.groupby('Pclass')['Survived'].apply(lambda x: x.astype(int).mean()))
print(titanic_data.groupby('Embarked')['Survived'].apply(lambda x: x.astype(int).mean()))
titanic_data.to_csv('titanic_cleaned.csv', index=False)
print("\n Cleaned dataset saved as 'titanic_cleaned.csv'.")
print("\n Task Completed Successfully.")
print(" Review visualizations and insights above.")
print(" Use 'titanic_cleaned.csv' for further modeling or analysis.")
```
